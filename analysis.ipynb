```python
maps={'southwest':0, 'southeast':1, 'northwest':2, 'northeast':3}
df['region']=df['region'].map(maps)
```

```python
# fix the sex column
maps={'male':0, 'female':1}
df['sex']=df['sex'].map(maps)
```

```python
# fix the smoker column
maps={'yes':0, 'no':1}
df['smoker']=df['smoker'].map(maps)
```

```python
df.isnull().sum()
```

```python
# Create new features BMI Category (Underweight/Normal/Overweight/Obese),Age Group, Interaction: smoker × BMI
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'
df['BMI_Category'] = df['bmi'].apply(bmi_category)

def age_group(age):
    if age < 30:
        return 'Young'
    elif 30 <= age < 50:
        return 'Middle-aged'
    else:
        return 'Senior'
df['Age_Group']=df['age'].apply(age_group)
df['Smoker_BMI_Interaction']=df['smoker']*df['bmi']
df
```
